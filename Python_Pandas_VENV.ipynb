# Handling Missing Values

```python
import numpy as np
import pandas as pd

# %%
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Adam', 40, np.nan, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['lynda', 42, 85, np.nan],
               [np.nan, 42, 85, np.nan],
               ['Jenny', np.nan, 100, 25632],
               ['Kenn', np.nan, 110, 25632],
                ['Aly', np.nan, 90, 25582],
               ['John', 41, 85, 1528000]]

df = pd.DataFrame(employee_list, columns=['Name', 'Age', 'Weight', 'Salary'])

# %%
rows, cols = df.shape

print(f'Rows: {rows}')
print(f'Cols: {cols}')

# %%
df

# %%
# Provide a summarization of missing values per column
df.isnull().sum()

# %%
# Be wary of missing values, if you have a small dataset, removing values can dramatically alter your analysis
# is missing data ROWS is less than 5% of all data it is safe to remove and will have little effect on analysis
dropped_df = df.dropna()
# when dropping na, you remove the entire row

print(dropped_df)

# %%
filled_df = df.fillna('None')

filled_df

# %%
filled_df.isnull().sum()

# %%
# Slicing and imputing columns at the same time
specific_cols = df[['Weight', 'Age']].fillna('Pending')

specific_cols

# %%
df2 = df.fillna(value={'Weight': 'Pending', 'Age': 'Unknown', 'Salary': '0.0', 'Name': 'Verification Pending'})

df2

# %%
technologies = {
    'Courses':["Spark",np.nan,"Hadoop","Python","pandas",np.nan,"Java"],
    'Fee' :[20000,25000, np.nan,22000,24000,np.nan,22000],
    'Duration':[np.nan,'40days','35days', np.nan,'60days','50days','55days'],
    'Discount':[1000,np.nan,1500,np.nan,2500,2100,np.nan]
              }

tech = pd.DataFrame(technologies)

# %%
rows, cols = tech.shape

print('Rows', rows)
print('Cols', cols)

# %%
tech.isnull().sum()

# %%
tech2 = tech.replace(np.nan, '')

tech2

# %%
tech3 = tech.replace({'Courses': np.nan, 'Fee': np.nan}, 'PENDING')

tech3

# %%
tech.fillna('')

tech

# %%
emp_df = pd.read_csv('data/employee.csv')

emp_df

# %%
emp_df.isnull().sum()

# %%
emp_df = emp_df.fillna(value={'Name': 'Verification Pending', 'Age': 'Unknown', 'Weight': 'pending', 'Salary': 0.0})

emp_df
```
