# Analyzing Monthly Expense

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ExpenseCategories = ['Groceries', 'Utilities', 'Transportation', 'Dining Out', 'Entertainment']
AmountSpent = [500, 300, 200, 400, 250]
colors = ['blue', 'orange', 'green','red', 'purple']
plt.barh(ExpenseCategories, AmountSpent, color=colors)
plt.xlabel('Amount Spent')
plt.ylabel('Expense Categories')
plt.title('Monthly Expenses Distribution')
plt.show()
```

```python


plt.figure(figsize=(10, 8))
plt.barh('Groceries', 500, color='blue', height=0.5)
plt.barh('Utilities', 300, color='orange', height=0.5)
plt.barh('Transportation', 200, color='green', height=0.5)
plt.barh('Dining Out', 400, color='red', height=0.5)
plt.barh('Entertainment', 250, color='purple', height=0.5)

plt.xlabel('Amount Spent')
plt.ylabel('Expense Categories')
plt.title('Montly Expenses Distribution')
plt.show()

```
