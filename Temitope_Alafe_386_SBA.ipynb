# Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read CSV file

```python
df = pd.read_csv("woocommerce-product-export.csv")
```

```python
print("Dataset Loaded Sucessfully")
```

# Print total_profit and month_number columns

```python
print("\nTOTAL PROFIT AND MONTH NUMBER:")
print(df[['month_number', 'total_profit']])
```

# Bar Plot for total profit per month

```python
months = df['month_number']
profit = df['total_profit']

plt.figure(figsize=(8,5))
plt.bar(months, profit)
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")

plt.show()
```

# Line Plot for total profit per month

```python
plt.figure(figsize=(8,5))

plt.plot(
    months,
    profit,
    linestyle='dotted',
    color='red',
    marker='o',
    markerfacecolor='black',
    linewidth=3,
    label='Profit Data'
)

plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")
plt.legend(loc='lower right')
plt.show()
```

# Multi-Line Plot for all product sales

```python
plt.figure(figsize=(10,6))

plt.plot(df['month_number'], df['facecream'], marker='o', label='Face Cream Sales Data')
plt.plot(df['month_number'], df['facewash'], marker='o', label='Face Wash Sales Data')
plt.plot(df['month_number'], df['toothpaste'], marker='o', label='Toothpaste Sales Data')
plt.plot(df['month_number'], df['bathingsoap'], marker='o', label='Bathingsoap Sales Data')
plt.plot(df['month_number'], df['shampoo'], marker='o', label='Shampoo Sales Data')
plt.plot(df['month_number'], df['moisturizer'], marker='o', label='Moistturizer Sales Data')

plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Sales Data")
plt.legend()
plt.show()





```

# Scatter Plot for bathingsoap sales data

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['month_number'],
    df['bathingsoap'],
    label='bathing soap Sales data'

)

plt.xlabel("Month Number")
plt.ylabel("Number of units Sold")
plt.title("bathingsoap Sales data")
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

```python
height = [121.9, 124.5, 129.5, 134.6, 139.7, 147.3, 152.4, 157.5, 162.6]
weight = [19.7, 21.3, 23.5, 25.9, 28.5, 32.1, 35.7, 39.6, 43.2]

plt.figure(figsize=(8, 5))
plt.plot(
    weight,
    height,
    linestyle='dashdot',
    color='green',
    marker='o',
    markersize=10,
    markerfacecolor='green',
    markeredgecolor='green',
    label='Height and Weight Date'

)
plt.xlabel("Weight in kg")
plt.ylabel("Height in cm")
plt.title("Average weight with respect to average height")
plt.legend(loc='lower right')
plt.show()
```
