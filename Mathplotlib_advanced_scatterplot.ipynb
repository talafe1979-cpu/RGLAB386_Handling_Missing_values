# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
plt.scatter(x, y, color='red', s=90)
plt.grid()
plt.show()
```

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, color='green', s=sizes, alpha=0.4)
plt.grid()
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10* np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
AgeGroupOne = range(40, 70, 1)
Purchase1=np.abs(np.random.randn(30) * 20)

plt.scatter(AgeGroupOne, Purchase1, label='40-69 Years Old', alpha=0.6, s=200)
plt.scatter(AgeGroupOne, Purchase1, color='green')

AgeGroupTwo=range(20, 40, 1)
Purchase2=np.abs(np.random.rand(20) * 40)

plt.scatter(AgeGroupTwo, Purchase2, label='20-39 Years Old', alpha=0.6, s=200)
plt.scatter(AgeGroupTwo, Purchase2, color='green')

plt.legend()
plt.xlabel('Age')
plt.ylabel('Purchases')
plt.title('Purchases by Age')
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.rand(100)
y = rng.rand(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.figure(figsize=(10, 7))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.title('Scatter Chart')
plt.xlabel('Rnadom data for x')
plt.ylabel('Rnadom data for y')
plt.colorbar()
plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv')
df.to_csv('housing_data.csv')
df.head()
```

```python
rows, cols = df.shape
print(f'Rows: {rows}')
print(f'Cols: {cols}')
```

```python
df.tail()
df.info()
df.isnull().sum()
df.duplicated().sum()
print(df.describe())

```

```python
plt.figure(figsize=(12, 8))
plt.scatter(
    x=df['longitude'],
    y=df['latitude'],
    alpha=0.4,
    s=df['population'] /100,
    c=df['median_house_value'],
    cmap=plt.get_cmap('jet')
)
    
plt.colorbar(label='Median Housing Value')
plt.title('Housing Data Scatter Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(['Size = Population'])
plt.show()
    


```
