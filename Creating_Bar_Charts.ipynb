# Creating Bar Charts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
Student = [90, 60, 130, 190]

plt.bar(CoursesName, Student, color='green', width=0.5, zorder=2)
plt.grid(True, axis='y', zorder=0)
plt.xlabel('Course Name')
plt.ylabel('Student Reg Number')
plt.title('Number of Student Per Course')
plt.show()
```

```python
plt.barh(CoursesName, Student, height=0.5)
plt.xlabel('Student Reg Number')
plt.ylabel('Course Name')
plt.title('Number of Student Per Course')
plt.show()
```

```python


male = [90, 60, 130, 600]
female = [100, 30, 160, 690]

plt.bar(CoursesName, male, width=0.5, label='Male')
plt.bar(CoursesName, female, bottom=male, width=0.5, label='Female')
plt.title('Number of Students Per Course')
plt.xlabel('Course Name')
plt.ylabel('Number of students')
plt.legend()
plt.show()

```

```python
years = ['2016', '2017', '2018', '2019']
rice_price = [57, 50, 47, 30]
wheat_price = [43, 50, 53, 70]

plt.figure(figsize=(8, 6))
plt.bar(years, rice_price, label='Rice', width=0.6)
plt.bar(years, wheat_price, label='Wheat', bottom=rice_price, width=0.6)
plt.title('Preference of Wheat/Rice Per Year', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Preference (%)', fontsize=15)
plt.legend(loc=2)
plt.show()
```

```python


mustard =  np.array([10, 16, 13, 15])

plt.figure(figsize=(8, 6))

plt.bar(years, rice_price, label='Rice', width=0.5)
plt.bar(years, wheat_price, label='Wheat', bottom=np.array(rice_price), width=0.5)
plt.bar(years, mustard, label='Mustad', bottom=np.array(rice_price) + np.array(wheat_price), width=0.5 )

plt.title('Rice, Wheat, Mustard By Year', fontsize=20)
plt.xlabel('Years')
plt.ylabel('Grains Frequency')
plt.legend()

plt.show()

```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
StudentGroup1 = [90, 60, 130, 600]
StudentGroup2 = [100, 30, 160, 690]

bar_width = 0.35
x = np.arange(len(CoursesName))

plt.bar(x - bar_width/2, StudentGroup1, label='Student Group 1', width=bar_width)
plt.bar(x + bar_width/2, StudentGroup2, label='Student Group 2', width=bar_width)
plt.xticks(x, CoursesName, rotation=45)
plt.legend()

plt.show()
```

```python



data = {'CoursesName': ['C Programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'],
        'Student Group 1': [90, 60, 130, 600],
        'Student Group 2': [100, 30, 160, 690]}

df = pd.DataFrame(data)

width = 0.35
x = df.index

plt.bar(x - width/2, df['Student Group 1'], width=width, label='Student Group 1')
plt.bar(x + width/2, df['Student Group 2'], width=width, label='Student Group 2')
plt.xticks(x, df['CoursesName'], rotation=45)
plt.title('NUmber of Students Per Course')
plt.ylabel('Number of Students')
plt.xlabel('Course Name')
plt.legend()
plt.show()

```
