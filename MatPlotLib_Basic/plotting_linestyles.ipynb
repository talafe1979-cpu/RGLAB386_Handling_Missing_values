# R-GLAB 386.5.1 Plotting Line Styles

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])
```

```python
plt.plot(xpoint, ypoint, linestyle='-')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle='--')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle='dashdot')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle=':')
plt.show()
```
