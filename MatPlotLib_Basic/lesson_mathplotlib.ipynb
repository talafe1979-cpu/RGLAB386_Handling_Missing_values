# Data Visualization with MATHPLOTLIB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])
```

```python
plt.plot(xpoint, ypoint)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(xpoint, ypoint)
plt.show()
```
